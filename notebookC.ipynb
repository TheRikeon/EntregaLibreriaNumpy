El dataset simple solo tiene las observaciones de General government expenditure by function de cada país de la unión europea, de la union europea en si y la euro area. Los años también solo van del 2012 al 2021.

```python
# en este dataset la primera columna es la única con texto, así que para facilitar el trabajo la separo.
# Como tiene varios valores los separo para poder leerlos individualmente.
datos_col1=np.loadtxt("data/gov_10a_exp_page_tabular.tsv", delimiter="\t",usecols=0,dtype=str,skiprows=1)
cantidad_observaciones=len(datos_col1)
tipo_observaciones = np.vstack(np.chararray.split(datos_col1, sep =','))
```

```python
# Cargo el resto de datos con genfromtxt para rellenar con nan los missing values
nombres=np.genfromtxt("data/gov_10a_exp_page_tabular.tsv", delimiter="\t",dtype=str,
                                skip_footer=cantidad_observaciones)
anyo_datos=np.char.strip(np.delete(nombres,0))
nombre_tipos=nombres[0].split(',')
datos_observaciones=np.genfromtxt("data/gov_10a_exp_page_tabular.tsv", delimiter="\t",usecols=np.arange(1,len(anyo_datos)+1),
                    encoding=None,skip_header=1)
```

```python
#Represento gráficamente la evolución del COFOG para España
idx_es=np.where(tipo_observaciones=='ES')
datos_es=datos_observaciones[idx_es[0]]
plt.bar(range(len(anyo_datos)),datos_es[0])
plt.xticks(range(len(anyo_datos)),anyo_datos,rotation=90)
plt.title("Evolution of Spain's COFOG")
plt.xlabel('Year')
plt.ylabel('Expenditure')
plt.show()
```

```python
plt.bar(range(len(datos_observaciones[:,5])),datos_observaciones[:,5])
plt.xticks(range(len(datos_observaciones[:,5])),tipo_observaciones[:,5],rotation=90)
plt.title("COFOG in the year "+anyo_datos[5])
plt.xlabel('Institution')
plt.ylabel('Expenditure')
plt.show()
```
